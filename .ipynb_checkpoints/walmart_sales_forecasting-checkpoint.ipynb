# Loads and preprocesses the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv', parse_dates=['Date'], index_col='Date')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.sort_index()
```

```python
df.index = pd.DatetimeIndex(df.index).to_period('W')
```

```python
weekly_sales = df.groupby('Date').sum()['Weekly_Sales']
```

```python
weekly_sales.index = weekly_sales.index.to_timestamp()
```

```python
df.head()
```

# Data Visualization

```python
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales, label='Weekly Sales')
plt.title('Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(weekly_sales, model='additive')
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

# Splits the data into training and test sets

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error
```

```python
train = weekly_sales[:'2012-01-01']
test = weekly_sales['2012-01-02':]
```

# Fits an ARIMA model on the training data and forecasts on the test data.

```python
# Fit the ARIMA model
model = ARIMA(train, order=(5, 1, 0))
model_fit = model.fit()

# Make predictions
forecast = model_fit.forecast(steps=len(test))
test = test.to_frame()  # Convert to DataFrame
test['forecast'] = forecast
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train, label='Train')
plt.plot(test['Weekly_Sales'], label='Test')
plt.plot(test['forecast'], label='Forecast')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

# Evaluates the model using Mean Absolute Error (MAE)

```python
mae = mean_absolute_error(test['Weekly_Sales'], test['forecast'])
print(f'Mean Absolute Error: {mae}')
```

# Plots the historical sales along with the forecasted sales.

```python
final_model = ARIMA(weekly_sales, order=(5, 1, 0))
final_model_fit = final_model.fit()
future_forecast = final_model_fit.forecast(steps=12)  # Forecast for the next 12 periods

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales, label='Historical Sales')
plt.plot(pd.date_range(start=weekly_sales.index[-1], periods=12, freq='W'), future_forecast, label='Forecast')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
